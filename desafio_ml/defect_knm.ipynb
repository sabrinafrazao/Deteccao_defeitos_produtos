```python
import os
import pandas as pd

current_dir = os.getcwd()
path_file = os.path.join(current_dir, "defects_data.csv")
df = pd.read_csv(path_file, header=0)

df.head()
```

```python
df = df.drop(['defect_date', 'defect_id'], axis=1)
```

```python
df.shape
```

```python
df['severity'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

for col in df.columns:
    if df[col].dtype == object:
        encoded = LabelEncoder()
        encoded.fit(df[col])
        df[col] = encoded.transform(df[col])

df.head(10)
```

```python
X = df[:600].drop('severity', axis=1)
y = df[:600]['severity']
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

# Define the number of epochs
epochs = 200  # You can change this number to the desired number of epochs

# Initialize lists to store accuracy for each epoch
train_accuracies = []
val_accuracies = []
test_accuracies = []

for epoch in range(epochs):
    # Train the model
    knn_model.fit(X_train, y_train)
    
    # Evaluate accuracy on the training set
    y_train_pred = knn_model.predict(X_train)
    train_accuracy = accuracy_score(y_train, y_train_pred)
    train_accuracies.append(train_accuracy)
    
    # Evaluate accuracy using cross-validation
    val_scores = cross_val_score(knn_model, X, y, cv=5)
    val_accuracy = val_scores.mean()
    val_accuracies.append(val_accuracy)
    
    # Evaluate accuracy on the test set
    y_test_pred = knn_model.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_accuracies.append(test_accuracy)
    
    # Print the results for the current epoch
    print(f'Epoch {epoch + 1}/{epochs} - Train Accuracy: {train_accuracy:.2f} - Validation Accuracy: {val_accuracy:.2f} - Test Accuracy: {test_accuracy:.2f}')
```
