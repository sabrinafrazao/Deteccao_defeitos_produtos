```python
import os
import pandas as pd

current_dir = os.getcwd()
path_file = os.path.join(current_dir, "defects_data.csv")
df = pd.read_csv(path_file, header=0)
```

```python
from sklearn.preprocessing import LabelEncoder

df = df.drop(['defect_date', 'defect_id'], axis=1)

label_encoders = {}

categorical_columns = ['defect_type', 'defect_location', 'severity', 'inspection_method']

for column in categorical_columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le

df.head(20)
```

```python
from ydata_profiling import ProfileReport

profile = ProfileReport(df)
profile.to_file('profile_output.html')
```
